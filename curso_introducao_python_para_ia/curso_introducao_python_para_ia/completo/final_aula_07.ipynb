# Aula 7: Dando accesso a sites para o nosso bot

O nosso objetivo é construir um bot que tenha acesso a informações personalizadas, que podem estar distribuídas em diversas fontes, como: sites, vídeos de Youtube, pdfs, etc.

A partir de agora, vamos dar opções ao nosso bot, para que ele consiga acessar as informações externas que ele utilizará como base para a geração de respostas mais específicas.

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://asimov.academy/")
lista_documentos = loader.load()
print(lista_documentos)
```

```python
documento = ''
for doc in lista_documentos:
  documento = documento + doc.page_content
```

```python
import os
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate

api_key = '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'
os.environ['GROQ_API_KEY'] = api_key

chat = ChatGroq(model='llama-3.1-70b-versatile')
```

```python
template = ChatPromptTemplate.from_messages([
    ('system', 'Você é um assistente amigável chamado Asimo e tem acesso as seguinte informações para dar as suas respostas: {documentos_informados}'),
    ('user', '{input}')
])

chain = template | chat
resposta = chain.invoke({'documentos_informados': documento, 'input': 'Quais as trilhas disponíveis na Asimov?'})
```

```python
print(resposta.content)
```
