## Acessando o Youtube

E agora vamos mostrar o quão fácil é de acessar vídeos de youtube utilizando LangChain!

```python
from langchain_community.document_loaders import YoutubeLoader

url = 'https://www.youtube.com/watch?v=nLopLwffsLI&ab_channel=AsimovAcademy'

loader = YoutubeLoader.from_youtube_url(
    url,
    language=['pt']
)
lista_documentos = loader.load()
```

```python
documento = ''
for doc in lista_documentos:
  documento = documento + doc.page_content
print(documento)
```

```python
template = ChatPromptTemplate.from_messages([
    ('system', 'Você é um assistente amigável que possui as seguintes informações para formular uma resposta: {informacoes}'),
    ('user', '{input}')
])

chain_youtube = template | chat
resposta = chain_youtube.invoke({'informacoes': documento, 'input': 'Aprender a programar é difícil?'})
print(resposta.content)
```

```python
resposta = chain_youtube.invoke({'informacoes': documento, 'input': 'Segundo o autor do documento, há quanto tempo ele programa em Python?'})
print(resposta.content)
```
